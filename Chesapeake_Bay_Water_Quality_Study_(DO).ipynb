### **Analyze based on months/seasons**

```python
#create month column
df_final['Month'] = df_final['SampleDate'].dt.month
```

```python
#monthly avgs
monthly_avg = df_final.groupby('Month')['MeasureValue'].mean()
print(monthly_avg)
```

```python
#show monthly avgs per year
monthly_by_year = df_final.groupby(['Year', 'Month'])['MeasureValue'].mean()

monthly_by_year = monthly_by_year.unstack()
print(monthly_by_year)
```

```python
monthly_by_year.T.plot()
```

```python
import pandas as pd
import plotly.express as px

# If monthly_by_year is already Year x Month, keep this
plot_df = monthly_by_year.reset_index().melt(
    id_vars='Year',
    var_name='Month',
    value_name='Average_DO'
)

plot_df['Month'] = plot_df['Month'].astype(int)
plot_df['Year'] = plot_df['Year'].astype(int)

fig = px.line(
    plot_df,
    x='Month',
    y='Average_DO',
    color='Year',
    markers=True,
    hover_data={
        'Year': True,
        'Month': True,
        'Average_DO': ':.2f'
    },
    title='Average Monthly Dissolved Oxygen in Chesapeake Bay by Year',
    labels={
        'Month': 'Month',
        'Average_DO': 'Dissolved Oxygen (mg/L)'
    }
)

fig.add_hline(
    y=5,
    line_dash='dash',
    annotation_text='Stress Threshold (~5 mg/L)',
    annotation_position='bottom right'
)

fig.update_layout(
    xaxis=dict(
        tickmode='array',
        tickvals=list(range(1, 13)),
        ticktext=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                  'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    ),
    legend_title_text='Year',
    hovermode='closest'
)

fig.show()

fig.write_html("monthly_do_plot.html")
```

```python
coverage = df.groupby('Year')[['Station', 'CBSeg2003', 'SampleDate']].nunique()
print(coverage.to_string())
```

The apparent increase in dissolved oxygen levels in 2025 is likely influenced by reduced spatial and temporal sampling. Compared to previous years, 2025 includes significantly fewer monitoring stations (28 vs ~49), fewer monitoring segments (9 vs 14), and fewer sampling dates. This reduced coverage may exclude lower-oxygen regions of the bay, leading to an artificially elevated average DO. Therefore, the 2025 increase should be interpreted cautiously and may not reflect a true system-wide improvement.
